```python
def stdev(nums):
    avg = sum(nums)/len(nums)
    diff = 0
    for n in nums:
        diff += (n - avg) ** 2
    return (diff / (len(nums) - 1)) ** 0.5
    
stdev(df.Height)
```

```python
def campaign_name(row):
    url = row['link']
    parsed = parse.urlsplit(url)
    params = parse.parse_qs(parsed.query)
    return params['utm_campaign'][0]

campaign_name(stats.loc[1])
```

```python
stats['campaign'] = stats.apply(campaign_name, axis = 1)
stats.head()
```

```python
def power_up(row, n):
    return row['effect'] ** n
```

```python
stats['power_up'] = stats.apply(power_up, n=2, axis = 1)
stats.head()
```
