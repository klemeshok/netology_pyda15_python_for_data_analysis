Рассчитаем базовые метрики для количественных величин и кратко опишем результаты:

```python
print('Средняя ректальная температура:', selected_df.rectal_temperature.mean())
print('Медиана ректальной температуры:', selected_df.rectal_temperature.median())
print('Мода ректальной температуры:', selected_df.rectal_temperature.mode()[0])
print('Стандартное отклонение ректальной температуры:', selected_df.rectal_temperature.std())
print('Дисперсия ректальной температуры:', selected_df.rectal_temperature.var())
print('Первый квартиль ректальной температуры:', selected_df.rectal_temperature.quantile(0.25))
print('Третий квартиль ректальной температуры:', selected_df.rectal_temperature.quantile(0.75))
```

Температурные средние очень плотны друг к другу, а отклонения незначительны. Так и должно быть: в более широких температурных диапазонах теплокровные живые существа не выживают.

```python
print('Средний пульс:', selected_df.pulse.mean())
print('Медиана пульса:', selected_df.pulse.median())
print('Мода пульса:', selected_df.pulse.mode()[0])
print('Стандартное отклонение пульса:', selected_df.pulse.std())
print('Дисперсия пульса:', selected_df.pulse.var())
print('Первый квартиль пульса:', selected_df.pulse.quantile(0.25))
print('Третий квартиль пульса:', selected_df.pulse.quantile(0.75))
```

Т.к. медиана пульса меньше его среднего арифметического, большинство лошадей имеет пульс меньше среднего. Значит среднее выростает за счет некоторого количества пиковых значений, которые могут претендовать на статус выбросов. Кроме этого, т.к. данные имеют большой разброс, а мода сильно удалена от других средних, можно предположить, что здесь наблюдается ситуация, когда есть много различных значений, отдельные из которых могут повторяться, но относительно небольшое количество раз. При таком распределении значений мода не очень показательна, т.к. ее значение  получилось в какой-то мере случайно (модой легко могло быть значение в совершенно другой части совокупности). Проверим эту гипотезу:

```python
# Гипотеза подтверждается.
selected_df.pulse.value_counts().head()
```

```python
print('Средний гематокрит:', selected_df.packed_cell_volume.mean())
print('Медиана гематокрита:', selected_df.packed_cell_volume.median())
print('Мода гематокрита:', selected_df.packed_cell_volume.mode()[0])
print('Стандартное отклонение гематокрита:', selected_df.packed_cell_volume.std())
print('Дисперсия гематокрита:', selected_df.packed_cell_volume.var())
print('Первый квартиль гематокрита:', selected_df.packed_cell_volume.quantile(0.25))
print('Третий квартиль гематокрита:', selected_df.packed_cell_volume.quantile(0.75))
```

Теперь подумаем, как заполнить пропуски в столбце 'abdominal_distension'. Для начала посмотрим, как распределились непустые значения помня, что пустых осталось 52 (296-244):

```python
selected_df.abdominal_distension.value_counts()
```

Хотя abdominal_distension (вздутие живота) - значение категориальное, провести замену пропусков модой здесь мы не можем, т.к. пропусков много, и такая замена, увы, приведет к существенному искажению данных. Но мы помним из описания данных, что у лошадей со вздутыми животами, вероятно, будет наблюдаться повышенная болезненность. И у них также снижена перистальтика кишечника. Кроме этого, лошадям с сильным вздутием живота, вероятно, потребуется операция только для устранения вздутия. Исходя из этих знаний можно было бы заполнить пропуски в показателях вздутия на основании данных других колонок. Но из перечисленных колонок мы включили в наш датафрейм только данные по хирургии. Поэтому будем отталкиваться от них, выполнив замену пропусков модой 'abdominal_distension', для каждого из значений 'surgery'. Это тоже приведет к искажению данных, но в меньшей степени.

```python
# Узнаем, какова же мода у 'abdominal_distension' при каждом из значений 'surgery':

print(selected_df[selected_df.surgery == 1].abdominal_distension.mode()[0])
print(selected_df[selected_df.surgery == 2].abdominal_distension.mode()[0])
```

```python
# Напишем функцию для замены

def replace_with_mode(row):
    '''Функция заменяет пустые значения в колонке 'abdominal_distension', в зависимости
    от значений  в колонке'surgery' '''
    
    if row['surgery'] == 1 and pd.isna(row['abdominal_distension']):
        return 3
    elif row['surgery'] == 2 and pd.isna(row['abdominal_distension']):
        return 1
    else:
        return row['abdominal_distension']
```

```python
# Применим функцию к датафрейму

selected_df['abdominal_distension'] = selected_df.apply(replace_with_mode, axis = 1)
```
