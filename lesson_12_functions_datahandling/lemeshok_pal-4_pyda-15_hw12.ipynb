#### Задание 2
<div class="alert alert-block alert-success">
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'. Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

    geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }

Результат классификации запишите в отдельный столбец region.

</div>

---

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

keywords = pd.read_csv('keywords.csv')
```

```python
def define_region(keyword):
    """Функция принимает на вход поисковый запрос и проверяет, указан ли в нем город, принадлежащий к заданному региону.
    Если город указан - функция возвращает название региона. Если не указан, функция возвращает 'undefined'.
    """
    words = keyword.lower().split(' ')
    for word in words:
        for key, value in geo_data.items():
            if word in value:
                return key
    return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(define_region)
keywords_with_region = keywords[ keywords.region != 'undefined']
keywords_with_region.head()
```

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска

```python
# Начнем именно с этого пункта, т.к. функция нам потом понадобится для выполнения остальных.

# Кроме этого, чтобы не искажать последующую статистику, изменим условие задачи:
# - если ни один из номеров года не встретился в названии фильма, то возвращаеся не '1900', а '0'.

def production_year(text):
    """Функция принимает на вход текст и проверяет, указан ли там год в интервале 1800-2020.
    Если указан - возвращает этот год. Если не указан - возвращает '0'.
    """
    import re
    try:
        if re.search(r'[\(18\d{2}\) | \(19\d{2}\) | \(200\d}\) | \(201\d\) | \(2020\)]', text):
            year = re.search(r'\(\d{4}\)', text)[0][1:5]
            return year
    except:
        return '0'
```

```python
# Запишем в переменную years список всех реальных годов, в которые снимались фильмы из нашей базы.

years_np = movies['title'].apply(production_year).unique()
years_np

# Вопросы:

# 1. На выходе получился тип данных numpy.ndarray, хотя мы не импортировали библиотеку Numpy
#    и работали с методами Pandas. Почему так?

# 2. В years_np как-то пробралось значение None. Константин, подскажите, пожалуйста, как такое может быть.
#    Ломаю голову, понять не могу. Вроде бы функция production_year это исключает, но по факту, увы, нет. 
```

```python
# преобразуем years_np в список, исключим значения None и '0', отсортируем полученный результат для красоты:
years = list(years_np)
years.remove(None)
years.remove('0')
years = sorted(years)
years
```

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга.

```python
# посчитаем средний рейтинг каждого фильма
mean_rating = ratings.groupby('movieId').mean().reset_index()[['movieId', 'rating']]
mean_rating.head()
```

```python
# отсортируем фильмы по movieId, чтобы потом к этому датафрейму добавить средний рейтинг каждого фильма
movies_sorted_by_id = movies.sort_values('movieId')
movies_sorted_by_id.head()
```

```python
# добавим в таблицу с фильмами их средний рейтинг
movies_sorted_by_id['rating'] = mean_rating['rating']
movies_sorted_by_id.loc[100:105]
```

```python
# создаем датафрейм year_rating, содержащий средний рейтинг фильмов каждого года
year_rating = (movies_sorted_by_id.groupby('year').mean().reset_index()[['year', 'rating']]
               .sort_values('rating', ascending=False))
year_rating.head(30)

# И убеждаемся, что гипотеза “раньше снимали настоящее кино, не то что сейчас” - подтвердилась.
```
