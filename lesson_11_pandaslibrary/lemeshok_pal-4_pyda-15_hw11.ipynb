#### Задание 1
<div class="alert alert-block alert-success">
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

</div>

---

```python
movies = pd.read_csv(r'D:\OneDrive\!Netology\!Netology_Pal-4\01_Python_for_data_analysis\Databases\ml-latest\movies.csv')
ratings = pd.read_csv(r'D:\OneDrive\!Netology\!Netology_Pal-4\01_Python_for_data_analysis\Databases\ml-latest\ratings.csv')

# создаем датафрейм, содержащий только оценки 5.0
ratings_5 = ratings[ ratings.rating == 5.0]

# находим: Id фильма с наибольшим количеством оценок 5.0, а также количество таких оценок
winner_id_score = dict(ratings_5.movieId.value_counts().head(1))
for key, value in winner_id_score.items():
    winner_id = key
    winner_score = value

# находим название фильма с наибольшим количеством оценок 5.0
winner_id_name = dict(movies[ movies.movieId == winner_id ]['title'])
for value in winner_id_name.values():
    winner_name = value

# выводим результаты
print(f'Больше всего оценок 5.0, а именно {winner_score}, - было выставлено фильму {winner_name}.')
```

#### Задание 2
<div class="alert alert-block alert-success">
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

</div>

---

```python
data = pd.read_csv('power.csv')

# создаем датафрейм, содержащий только данные, соответствующие условиям задачи
filtered_data = data[ ((data['country'] =='Latvia') | (data['country'] =='Lithuania') | (data['country'] =='Estonia')) &
                    ((data['category'] == 4) | (data['category'] == 12) | (data['category'] == 21)) &
                     (data['year'] >= 2005) & (data['year'] <= 2010) & (data['quantity'] > 0)]

# считаем сумму в колонке filtered_data['quantity'] и выводим результат на экран
print(f'Cуммарное потребление стран Прибалтики по категориям 4, 12 и 21 за период с 2005 по 2010 года \
без учета отрицательных зачений потребления составляет {filtered_data.quantity.sum()}.')
```
